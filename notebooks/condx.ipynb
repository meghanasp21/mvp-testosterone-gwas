### Get conditional X hits

```python
afr_condx=pd.read_csv("../data/conditional-x/afr.extract",header=None)
afr_condx["chr"]=afr_condx[0].str.split(":").str[0]
afr_condx["bp1"]=afr_condx[0].str.split(":").str[1]
afr_condx["bp1"]=afr_condx["bp1"].astype(int)
afr_condx["bp2"]=afr_condx["bp1"]+1
```

```python
eur_condx=pd.read_csv("../data/conditional-x/eur.extract",header=None)
eur_condx["chr"]=eur_condx[0].str.split(":").str[0]
eur_condx["bp1"]=eur_condx[0].str.split(":").str[1]
eur_condx["bp1"]=eur_condx["bp1"].astype(int)
eur_condx["bp2"]=eur_condx["bp1"]+1
```

```python
his_condx=pd.read_csv("../data/conditional-x/his.extract",header=None)
his_condx["chr"]=his_condx[0].str.split(":").str[0]
his_condx["bp1"]=his_condx[0].str.split(":").str[1]
his_condx["bp1"]=his_condx["bp1"].astype(int)
his_condx["bp2"]=his_condx["bp1"]+1
```

```python
afr_condx[["chr","bp1","bp2"]].to_csv("../data/conditional-x/genome-decoration/afr.condx.bed",header=None,index=None,sep="\t")
eur_condx[["chr","bp1","bp2"]].to_csv("../data/conditional-x/genome-decoration/eur.condx.bed",header=None,index=None,sep="\t")
his_condx[["chr","bp1","bp2"]].to_csv("../data/conditional-x/genome-decoration/his.condx.bed",header=None,index=None,sep="\t")
```

### Get sumstats

```python
direct="../data/conditional-x/sumstatsx/"
files=os.listdir(direct)
chunks=set([x.split("GWAS")[0] for x in files])
chunks=[x for x in chunks]
```

```python
for chunk in chunks:
    chunk_files=[x for x in files if chunk in x]
    
    compiled=pd.DataFrame()
    for file in chunk_files:
        df=pd.read_csv(direct+file,delimiter="\t")
        compiled=compiled.append(df)
    
    compiled.to_csv(direct+chunk+"GWAS.ADD",index=None,sep="\t")
        
```

### make locus zoom plots

```python
files=[x for x in files if x.endswith("ADD")]
condx=pd.DataFrame({"files":files})

condx["num"]=condx["files"]
condx["num"]=condx["num"].str.split(".").str[0]
condx["num"]=condx["num"].str.replace("afr","")
condx["num"]=condx["num"].str.replace("eur","")
condx["num"]=condx["num"].str.replace("his","")
condx["num"]=condx["num"].astype(int)

compiled_assoc=pd.DataFrame()

for group in ["eur","afr","his"]:
    df=pd.read_csv("../data/conditional-x/"+group+".assoc",header=None,delimiter="\t")
    df["files"]=condx[condx["files"].str.contains(group)].sort_values(by=["num"])["files"].tolist()
    df["group"]=group
    compiled_assoc=compiled_assoc.append(df)

compiled_assoc["marker"]="chr"+compiled_assoc[0].str.rsplit(":",2).str[0]

mp_group={"eur":"eur","afr":"afr","his":"amr"}
compiled_assoc["ref"]=compiled_assoc["group"].map(mp_group)

compiled_assoc["bp"]=compiled_assoc[0].str.split(":").str[1]
compiled_assoc["bp"]=compiled_assoc["bp"].astype(int)
```
